## Concatenate

```python
a = np.array([[1,2], [3,4]])
b= np.array([5, 6])
```

```python
#np.concatenate((a,b), axis=0)
# las dimensiones de a y b deben ser iguales
b = np.expand_dims(b, axis=0)
np.concatenate((a,b), axis=0)
```

```python
# np.concatenate((a,b), axis=1)
# para poder hacerlo por colunmas b debe ser una columna, para cumplir las dimensiones
np.concatenate((a,b.T), axis=1) # hacemos la transpuesta
```
