## Agregar o eliminar dimensiones

```python
vector = np.array([1,2,3], ndmin=10)
print(vector)
print(vector.ndim)
```

```python
expand = np.expand_dims(np.array([1,2,3]), axis=0) # 0 -> columnas 1-> filasç
print(expand)
expand.ndim
```

```python
print(vector, vector.ndim)
vector2 = np.squeeze(vector)# squeeze lo manda a la dimensionalidad que cree correcta
print(vector2, vector2.ndim)
```
