Cambio tipo de datos

```python
arr = np.array([1,2,3,4], dtype='float64')
print(arr)
print(arr.dtype)
```

```python
arr = np.array([1,2,3,4])
arr = arr.astype(np.float64)
print(arr)
print(arr.dtype)
```

```python
arr = np.array([0, 1,2,3,4])
arr = arr.astype(np.bool_)
print(arr)
print(arr.dtype)
# en boolean el 0 siempre es falso, mientras que el resto verdadero
```

```python
arr = np.array([0, 1,2,3,4])
arr = arr.astype(np.string_)
print(arr)
print(arr.dtype)
# En string los maneja como cadenas de caracteres
```

```python
arr = np.array(['0', '1','2','3','4'])
arr = arr.astype(np.int8)
print(arr)
print(arr.dtype)
# Es capaz de coger los numeros que son string a int
# pero falla si intenta convertir caracteres a numeros
```

El array de numpy debe tener todos los elementos del mismo tipo, si no fallar√°
